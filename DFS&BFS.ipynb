```python
def dfs(graph,start, visited=None):
  if visited is None:
    visited=set()
  visited.add(start)
  print(start,end=" ")
  for neighbor in graph[start]:
    if neighbor not in visited:
      dfs(graph,neighbor,visited)
graph={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}
print("DFS Traversal:")
dfs(graph,'A')
```

```python
from collections import deque
def bfs(graph,start):
    visited=set()
    queue=deque([start])
    while queue:
      node=queue.popleft()
      if node not in visited:
        print(node,end=' ')
        visited.add(node)
        queue.extend(graph[node])
graph={
    'A':['B','C'],
    'B':['D','E'],
    'C':['F'],
    'D':[],
    'E':['F'],
    'F':[]
}
print("BFS Traversal:")
dfs(graph,'A')
```
